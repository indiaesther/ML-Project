```python
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
comp=pd.read_excel('./data/model/company_all_token_210222.xlsx')
```

```python
comp.columns
```

```python
data=comp
```

```python
col_list = ['연구_okt', '연구_hannanum', '연구_krwordrank', '연구_kkma', 'okt_han_kkma']

for col in col_list:
    temp=data[col].tolist()
    word_list=[]
    for line in temp:
        line = line.split()
        for w in line:
            word_list.append(w)

    counts = Counter(word_list)
    tags = counts.most_common(150)

    wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf',
                          background_color='white',
                          width=800,
                          height=600)

    #print(dict(tags))

    cloud = wordcloud.generate_from_frequencies(dict(tags))
    print('======================%s WordCloud======================'%col)
    fig=plt.figure(figsize=(10, 8))
    plt.axis('off')
    plt.imshow(cloud)
    plt.show()
    fig.savefig('./data/wordcloud/%s_wordcloud.png'%col)
    print('======================저장완료======================')
```

# 지원사업 데이터

```python
support = pd.read_excel('./data/support_final_210222.xlsx')
```

```python
data=support
```

```python
col_list = ['all_word']

for col in col_list:
    temp=data[col].tolist()
    word_list=[]
    for line in temp:
        line = line.split()
        for w in line:
            word_list.append(w)

    counts = Counter(word_list)
    tags = counts.most_common(150)

    wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf',
                          background_color='white',
                          width=800,
                          height=600)

    #print(dict(tags))

    cloud = wordcloud.generate_from_frequencies(dict(tags))
    print('======================%s WordCloud======================'%col)
    fig=plt.figure(figsize=(10, 10))
    plt.axis('off')
    plt.imshow(cloud)
    plt.show()
    fig.savefig('./data/wordcloud/지원사업_%s_wordcloud.png'%col)
    print('======================저장완료======================')
```
